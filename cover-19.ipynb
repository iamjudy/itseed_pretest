```python
df= pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
df
```

```python
# 全球確診人數最高的一天


print("全球確診人數最高的一天：")

fliter = (df["location"] == "World")
y=df[fliter].new_cases.idxmax(skipna=True)
df.iloc[[y]]
```

```python
#找出確診國家中，12月新增確診成長率最高、最低的三個國家


df["date"] = pd.to_datetime(df["date"])  
mask = (df["date"] >= '2020-11-01') & (df["date"] <= '2020-11-30')
#print(df.loc[mask])

sector = df.loc[mask].groupby("iso_code")
nov = sector.new_cases.sum( )

maskk = (df["date"] >= '2020-12-01') & (df["date"] <= '2020-12-31')
sectorr = df.loc[maskk].groupby("iso_code")
dec = sectorr.new_cases.sum( )


import numpy as np

rate = (dec-nov).div(nov)
rate[np.isinf(rate)] = np.nan                               # 排除除數為零的情況
print("12月新增確診成長率最高的三個國家：")
print(rate.nlargest(5))                                       # 排前五名檢查同數值並列第三名的可能
print("\n")
print("12月新增確診成長率最低的三個國家：")
print(rate.nsmallest(5))

#驗算
#n=sector.get_group('MHL').new_cases.sum( )
#d=sectorr.get_group('MHL').new_cases.sum( )
#print(d-n/n)
```
